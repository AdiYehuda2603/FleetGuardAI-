## ğŸ”§ ×¤×•× ×§×¦×™×•×ª ×¢×–×¨ ×œ×¢×‘×¨×™×ª RTL

```python
def fix_hebrew_text(text):
    """×ª×™×§×•×Ÿ ×˜×§×¡×˜ ×¢×‘×¨×™×ª ×œ-RTL display"""
    if not text:
        return ""
    
    # ×× ×”×˜×§×¡×˜ ××›×™×œ ×¨×§ ×× ×’×œ×™×ª - ×”×—×–×¨ ×›××• ×©×”×•×
    if all(ord(char) < 128 or char in 'â‚ªÂ°Â±Ã—Ã·' for char in text.replace(' ', '')):
        return text
    
    # ×˜×™×¤×•×œ ×‘×¢×‘×¨×™×ª
    reshaped_text = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped_text)
    return bidi_text

def add_rtl_text(text_frame, text, font_size=18, bold=False, color=None):
    """×”×•×¡×¤×ª ×˜×§×¡×˜ RTL ×œ×©×§×•×¤×™×ª"""
    fixed_text = fix_hebrew_text(text)
    p = text_frame.add_paragraph()
    p.text = fixed_text
    p.alignment = PP_ALIGN.RIGHT
    p.font.size = Pt(font_size)
    p.font.bold = bold
    p.font.name = 'Arial'
    if color:
        p.font.color.rgb = color
    return p

print("âœ… ×¤×•× ×§×¦×™×•×ª RTL ××•×›× ×•×ª!")
```
