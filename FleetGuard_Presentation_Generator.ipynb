```python
# ×™×™×‘×•× ×¡×¤×¨×™×•×ª
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor
from bidi.algorithm import get_display
import arabic_reshaper
from google.colab import files

print("âœ… Libraries imported successfully!")
```

```python
# ×¤×•× ×§×¦×™×•×ª ×¢×–×¨ ×œ×¢×‘×¨×™×ª RTL
def fix_hebrew(text):
    """Convert Hebrew text to RTL display format"""
    if not text or all(ord(c) < 128 or c in 'â‚ªÂ°Â±Ã—Ã·' for c in text.replace(' ', '')):
        return text
    return get_display(arabic_reshaper.reshape(text))

def add_text(frame, text, size=18, bold=False, color=None):
    """Add RTL text to slide"""
    p = frame.add_paragraph()
    p.text = fix_hebrew(text)
    p.alignment = PP_ALIGN.RIGHT
    p.font.size = Pt(size)
    p.font.bold = bold
    p.font.name = 'Arial'
    if color:
        p.font.color.rgb = color
    return p

print("âœ… Helper functions ready!")
```
